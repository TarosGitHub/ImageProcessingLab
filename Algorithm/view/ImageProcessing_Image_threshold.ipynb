```python
import os
import sys
sys.path.append(os.path.join(os.getcwd(), '../..'))
import re
from PIL import Image as im
import matplotlib.pyplot as plt
import numpy as np
from Algorithm.src import ImageProcessing as ip

%matplotlib inline

SRC_IMAGE = '/SIDBA/Mono/LENNA.bmp'
DST_IMAGE = '/Algorithm/img/threshold_LENNA.bmp'

work_dir = re.search(r'^.*ImageProcessingLab', os.getcwd()).group()

image = ip.Image(work_dir + SRC_IMAGE)
threshold_image = image.threshold(100)
threshold_image.save(work_dir + DST_IMAGE)

plt.rcParams['figure.figsize'] = (15.0, 15.0)

before_img = im.open(work_dir + SRC_IMAGE)
before_img_list = np.asarray(before_img)
plt.subplot(1,2,1)
plt.imshow(before_img_list, cmap='gray', vmin = 0, vmax = 255)

after_img = im.open(work_dir + DST_IMAGE)
after_img_list = np.asarray(after_img)
plt.subplot(1,2,2)
plt.imshow(after_img_list, cmap='gray', vmin = 0, vmax = 255)

plt.show()
```
