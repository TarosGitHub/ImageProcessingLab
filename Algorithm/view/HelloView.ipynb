```python
import os
import re
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

work_dir = re.search(r'^.*ImageProcessingLab', os.getcwd()).group()

plt.rcParams['figure.figsize'] = (15.0, 15.0)

before_img = Image.open(work_dir + "/SIDBA/Color/Lenna.bmp")
before_img_list = np.asarray(before_img)
plt.subplot(1,2,1)
plt.imshow(before_img_list)

after_img = Image.open(work_dir + "/SIDBA/Mono/LENNA.bmp")
after_img_list = np.asarray(after_img)
plt.subplot(1,2,2)
plt.imshow(after_img_list, cmap='gray', vmin = 0, vmax = 255)

plt.show()
```
